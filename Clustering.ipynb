1. What is unsupervised learning in the context of machine learning?

-> Unsupervised Machine learning is the ml technique which is used on unlabeled data means data has not dependent variable or output only has independent features. using this feature we combine and make a clusters of same type of datapoints or similar types of datapoints.

2.  How does K-Means clustering algorithm work?

-> K-Means term said that K (no of groups) and Mean (average). K-Means is used to classify a new data point should be in which class.

steps of K-Means clustering is given below :
1. first we select the random k centroids for making the k clusters from the unlabeled data.
2. then we repetadely add new data points into this and every time we calculate the distance from the new data point to that nearest point and based on their nearest we assign that point into that cluster.
3. after that we recalculate the centroid (average of all points in that particular cluster).
4. this process is do recursively untill all the centroid is unchanged.
5. remember the centroid updating process is only done in training phase not in testing phase.

6. When would you use Silhouette Score in clustering?

-> Silhouette score is used when you want to check the clusters are made perfeclty sepreble to other clusters or not.

means the local cluster's data points are closely packed to each other and two different clusters are far away to each other.

the formula of silhouette score is
b - a/max(b-a)

where b = ocss (outer cluster sum of squared)
and a = wcss (within cluster sum of squared)

the curve is down as the increase of k value.

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
from sklearn.metrics import silhouette_score

X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=42)

ks = [2, 3, 4, 5]
scores = []

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.figure(figsize=(6,5))
plt.bar(ks, scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

data = load_breast_cancer()
X = data.data

inertia_values = []
K_range = range(2, 7)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(K_range, inertia_values, marker='o')
plt.title("K-Means Inertia for K=2 to 6")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(K_range)
plt.show()
```

36. Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding
noise)

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

data = load_wine()
X = data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=1.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X_scaled)

# In DBSCAN, noise is labeled as -1
n_clusters = len(set(y_dbscan)) - (1 if -1 in y_dbscan else 0)
print("Number of clusters (excluding noise):", n_clusters)
```

40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
scatter plot.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

digits = load_digits()
X = digits.data
y_true = digits.target

pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42)
y_clusters = kmeans.fit_predict(X_pca)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca[:,0], X_pca[:,1], X_pca[:,2], c=y_clusters, cmap='tab10', s=50)
ax.set_title("KMeans Clustering on Digits Dataset (PCA 3D)")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```
